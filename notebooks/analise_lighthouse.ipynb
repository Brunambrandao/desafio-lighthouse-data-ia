```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregamento dos dados
df_vendas = pd.read_csv('../data/vendas_2023_2024.csv')
df_prod = pd.read_csv('../data/produtos_raw.csv')

# Limpeza de duplicatas para evitar inflação de valores
df_prod_unique = df_prod.drop_duplicates(subset=['code']).copy()

# Agrupamento de vendas
vendas_resumo = df_vendas.groupby('id_product').agg({'total': 'sum', 'qtd': 'sum'}).reset_index()

# Cruzamento de dados
df_final = pd.merge(vendas_resumo, df_prod_unique[['code', 'price']], left_on='id_product', right_on='code', how='left')

# Cálculo do impacto financeiro (Câmbio 5.35)
taxa = 5.35
df_final['impacto_brl'] = df_final['total'] - (df_final['price'] * taxa * df_final['qtd'])

print(f'Impacto Total Consolidado: R$ {df_final["impacto_brl"].sum():,.2f}')
```

```python
# Visualização dos 10 maiores impactos negativos
top_10 = df_final.sort_values(by='impacto_brl', ascending=True).head(10)

plt.figure(figsize=(12, 6))
plt.bar(top_10['id_product'].astype(str), top_10['impacto_brl'].abs(), color='#d63031')
plt.title('Top 10 Produtos com Maior Impacto Financeiro Negativo (BRL)')
plt.ylabel('Prejuízo Acumulado (R$)')
plt.show()
```
